# 3. Check for the chances of getting diabetes if it is common in family, using the Diabetes Pedigree Function

```python
sns.violinplot(y=d['DiabetesPedigreeFunction'],x=d['Outcome'],data=d)
```

# 4. Check for the relationship with Insulin & Glucose and group the datapoints by Outcome.

```python
plt.figure()
sns.relplot(x="Glucose", y="Insulin", hue="Outcome", data=d ,style="Outcome",markers=True,col="Outcome")
plt.xticks(rotation=75)
```
